```python
import networkx as nx
!pip install PyPDF2
import PyPDF2
```

```python
import re

capitulos = []
# Define el patrón para buscar el nombre del personaje y la ubicación separados por números romanos
pattern = r"([A-Za-z\s]+)\s\b(?:[IVXLCDM]+)\b\s(.*?)(?=(?:North|Riverlands|Crownlands|Vale|Iron Islands|Reach|Stormlands|Dorne|Free Cities|Dothraki Sea|Far East|Slaver's Bay|Rhoyne|Beyond-the-Wall|Open oceans|$))"
# Abre el archivo PDF en modo lectura binaria ('rb')
with open('asoiaf_places.pdf', 'rb') as file:
    # Crea un objeto PDFReader
    pdf_reader = PyPDF2.PdfReader(file)
    # Obtén el número de páginas del PDF
    num_pages = len(pdf_reader.pages)

    # Itera sobre cada página y extrae su contenido de texto
    for page_num in range(2, 11):  # Suponiendo que las páginas relevantes están en este rango
        # Obtiene el objeto de la página
        page = pdf_reader.pages[page_num]

        # Extrae el texto de la página
        text = page.extract_text()

        # Busca coincidencias con el patrón en el texto
        matches = re.findall(pattern, text)

        # Imprime los resultados
        for match in matches:
            character = match[0].strip()  # Nombre del personaje
            location = match[1].strip()   # Ubicación
            capitulos.append((character, location))


        for i, tupla in enumerate(capitulos):
            if tupla[0] == 'Daenerys' and tupla[1] == '':
              capitulos[i] = (tupla[0], 'Dothraki Sea')
            elif tupla[0] == 'Arya' and tupla[1] == '':
              capitulos[i] = (tupla[0], 'Harrenhal')
            elif tupla[0] == 'Victarion' and tupla[1] == '':
              capitulos[i] = (tupla[0], 'Slaver\'s Bay')
            elif tupla[0] == 'Brienne' and tupla[1] == '':
              capitulos[i] = (tupla[0], 'Red Fork')
            elif tupla[0] == 'Tyrion' and tupla[1] == '':
              capitulos[i] = (tupla[0], 'Rhoyne')
```

```python
capitulos
```

```python
# Lista para almacenar los resultados
result = []

# Iterar sobre cada tupla en la lista original
for personaje, lugar in capitulos:
    if '/' in lugar:
        # Si hay una barra, dividir el lugar en dos partes
        lugar1, lugar2 = lugar.split(' / ')
        # Crear dos nuevas tuplas y agregarlas a la lista de resultados
        result.append((personaje, lugar1))
        result.append((personaje, lugar2))
    else:
        # Si no hay barra, agregar la tupla original a la lista de resultados
        result.append((personaje, lugar))

# Imprimir el resultado
for item in result:
    print(item)
```

```python
# Convertir la lista temporal a un conjunto para eliminar duplicados, luego volver a convertirlo a una lista
capitulos = list(set(result))
capitulos
```

```python
# Filtra las tuplas en la lista capitulos donde el segundo elemento está vacío
tuplas_vacias = [capitulo for capitulo in capitulos if capitulo[1] == '']

# Imprime las tuplas obtenidas
for tupla in tuplas_vacias:
    print(tupla)

len(capitulos)
```
